```python
# Define a function to get the chatbot's response
def get_response(user_input):
    # Convert user input to lowercase for case-insensitive matching
    user_input = user_input.lower()

    # Check for various queries and provide appropriate responses
    if "hello" in user_input:
        return "Hello! How can I assist you today?"
    elif "help" in user_input:
        return "Sure! What do you need help with?"
    elif "goodbye" in user_input:
        return "Goodbye! Have a great day!"
    else:
        return "I'm sorry, I didn't understand your query."

# Main interaction loop
while True:
    user_input = input("User: ")

    # Exit the loop if the user says goodbye
    if "goodbye" in user_input.lower():
        print("Bot: Goodbye! Have a great day!")
        break

    # Get the bot's response and print it
    response = get_response(user_input)
    print("Bot:", response)
```
